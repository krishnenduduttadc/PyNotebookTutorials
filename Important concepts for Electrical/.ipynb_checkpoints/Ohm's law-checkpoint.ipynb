```python
import numpy as np
import plotly.graph_objects as go

# Resistance range
R = np.linspace(0.1, 20, 200)

# Voltage values for slider steps
voltages = np.arange(1, 21, 1)  # 1V to 20V

# Prepare traces for each voltage level
traces = []
for V in voltages:
    I = V / R
    traces.append(go.Scatter(x=R, y=I, mode='lines', name=f'V={V}V', visible=False))

# Make the first trace visible
traces[0].visible = True

# Create the figure
fig = go.Figure(data=traces)

# Create slider steps
steps = []
for i, V in enumerate(voltages):
    step = dict(
        method="update",
        args=[{"visible": [False] * len(traces)},
              {"title": f"Ohm's Law: Current vs Resistance (Voltage = {V} V)"}],
        label=str(V)
    )
    step["args"][0]["visible"][i] = True  # Toggle i-th trace to visible
    steps.append(step)

# Create sliders object
sliders = [dict(
    active=0,
    currentvalue={"prefix": "Voltage: "},
    pad={"t": 50},
    steps=steps
)]

# Update layout with sliders
fig.update_layout(
    sliders=sliders,
    xaxis_title="Resistance (Ω)",
    yaxis_title="Current (A)",
    template="plotly_white",
    height=400
)

fig.show()
```
