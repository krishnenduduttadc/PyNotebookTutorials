# Lesson 26: Busbar Configurations in Substations

## What is a Busbar?

A **busbar** is a conductor (usually copper or aluminum) used for collecting electric power from incoming feeders and distributing to outgoing feeders. It's the backbone of a substation layout.

---

## Common Bus Configurations:

### 1. **Single Busbar Scheme**

- All circuits (incoming and outgoing) are connected to one bus.
- **Advantages**: Simple, cheap.
- **Disadvantages**: No redundancy â€” entire system shuts down if bus fails.



```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    d.config(unit=2.5)

    # Incoming 220kV line and breaker
    d += elm.SourceV().label('Incoming\n220kV')
    d += elm.Line().right()
    d += elm.Switch().label('CB-1', loc='top')  # Circuit Breaker
    d += elm.Line().right().length(2.5)

    # Busbar simulated with a horizontal line
    bus_start = d.here
    d += elm.Line().right().length(6)
    bus_end = d.here

    # Feeder A (downward)
    d += elm.Line().down().at((bus_start[0] + 1, bus_start[1]))
    d += elm.Switch().down().label('CB-A', loc='top')
    d += elm.Line().down()
    d += elm.ResistorIEC().down().label('Feeder A\n33kV', loc='bottom')

    # Feeder B (downward)
    d += elm.Line().down().at((bus_start[0] + 3, bus_start[1]))
    d += elm.Switch().down().label('CB-B', loc='top')
    d += elm.Line().down()
    d += elm.ResistorIEC().down().label('Feeder B\n33kV', loc='bottom')

    # Feeder C (downward)
    d += elm.Line().down().at((bus_start[0] + 5, bus_start[1]))
    d += elm.Switch().down().label('CB-C', loc='top')
    d += elm.Line().down()
    d += elm.ResistorIEC().down().label('Feeder C\n33kV', loc='bottom')
```
