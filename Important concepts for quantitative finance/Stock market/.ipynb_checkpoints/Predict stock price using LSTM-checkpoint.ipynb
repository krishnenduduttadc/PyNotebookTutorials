```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
# !pip install tensorflow scikit-learn
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
```

```python
#Load & Preprocess Data for Infosys share (LSTM)
```

```python
df = yf.download("INFY.NS", start="2015-01-01", end="2025-06-01")
df = df[['Close']]
df.head()
```

```python
 # Normalize Data
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

plt.plot(scaled_data)
plt.title("Normalized Close Price")
plt.show()
```

```python
 # Create Time-Series Dataset for LSTM
```

```python
def create_dataset(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i - window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

window_size = 60
X, y = create_dataset(scaled_data, window_size)

# Reshape for LSTM [samples, time_steps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Train-Test Split
```

```python
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
# Build and Train LSTM Model
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Plot Training Loss
```

```python
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# Predict and Inverse Transform
```

```python
predictions = model.predict(X_test)
predicted_prices = scaler.inverse_transform(predictions)
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
# Visualize Actual vs Predicted Prices
```

```python
plt.figure(figsize=(14, 6))
plt.plot(actual_prices, label='Actual Price')
plt.plot(predicted_prices, label='Predicted Price')
plt.title('LSTM Model - AAPL Price Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
#Multi-Step Forecast Until Sept 30, 2025
```

```python
from datetime import datetime, timedelta

# Predict forward to target date
latest_df = yf.download('INFY.NS', start='2015-01-01', end='2025-06-06')[['Close']]
latest_scaled = scaler.fit_transform(latest_df)
last_60_days = latest_scaled[-60:]
X_future = np.reshape(last_60_days, (1, 60, 1))
target_date = datetime(2025, 9, 30)
current_date = latest_df.index[-1]
predicted_prices = []

input_seq = last_60_days.copy()

while current_date < target_date:
    pred = model.predict(input_seq.reshape(1, 60, 1))
    predicted_prices.append(pred[0][0])
    
    # Append prediction and remove first value to maintain window size
    input_seq = np.append(input_seq, pred)[1:].reshape(60, 1)
    current_date += timedelta(days=1)
    
    # Skip weekends
    while current_date.weekday() >= 5:
        current_date += timedelta(days=1)

# Inverse scale predicted values
predicted_prices_real = scaler.inverse_transform(np.array(predicted_prices).reshape(-1, 1))

# Final prediction for Sept 30
print(f"Predicted stock price on 30th Sep 2025: {predicted_prices_real[-1][0]:.2f}")
```

```python
# Predict Next Day Price (1-Step Forecast)
```

```python
future_pred_scaled = model.predict(X_future)
future_pred = scaler.inverse_transform(future_pred_scaled)
print("Predicted price for next day (1-step):", future_pred[0][0])
```
