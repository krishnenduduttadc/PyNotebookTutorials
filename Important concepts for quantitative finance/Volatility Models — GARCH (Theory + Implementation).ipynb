## GARCH(1,1) Model:

The conditional variance \( \sigma_t^2 \) evolves as:

\[
\sigma_t^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2
\]

Where:

- \( \epsilon_t \): return shock at time \( t \) (innovation)
- \( \sigma_t^2 \): conditional variance at time \( t \)
- \( \omega > 0 \): constant term
- \( \alpha \geq 0 \): impact of recent squared shocks (ARCH term)
- \( \beta \geq 0 \): persistence of past variance (GARCH term)

---

## Key Characteristics:

- Captures volatility clustering: periods of high volatility tend to cluster.
- Useful for risk management, option pricing, and forecasting volatility.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Parameters
omega = 0.00001
alpha = 0.1
beta = 0.85
T = 1000

# Initialize arrays
sigma2 = np.zeros(T)
epsilon = np.zeros(T)

# Starting values
sigma2[0] = omega / (1 - alpha - beta)
epsilon[0] = np.sqrt(sigma2[0]) * np.random.normal()

# Simulate GARCH(1,1)
for t in range(1, T):
    sigma2[t] = omega + alpha * epsilon[t-1]**2 + beta * sigma2[t-1]
    epsilon[t] = np.sqrt(sigma2[t]) * np.random.normal()

# Plot
plt.figure(figsize=(10,6))
plt.plot(np.sqrt(sigma2), label='Conditional Volatility (Ïƒ_t)')
plt.title('Simulated GARCH(1,1) Conditional Volatility')
plt.xlabel('Time')
plt.ylabel('Volatility')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
