## Vasicek Short-Rate Model:

\[
dr_t = a(b - r_t) dt + \sigma dW_t
\]

Where:

- \( r_t \): short-term interest rate at time \( t \)
- \( a \): speed of mean reversion
- \( b \): long-term mean rate
- \( \sigma \): volatility of interest rates
- \( dW_t \): Wiener process (Brownian motion)

---

### Interpretation:

- The rate tends to revert to mean \( b \) at speed \( a \).
- Random shocks \( \sigma dW_t \) introduce uncertainty.
- It is an Ornstein-Uhlenbeck process â€” Gaussian and mean reverting.

---

### Applications:

- Pricing


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# Vasicek parameters
a = 0.15       # speed of mean reversion
b = 0.05       # long-term mean
sigma = 0.01   # volatility
r0 = 0.03      # initial rate
T = 1          # time horizon (1 year)
dt = 1/252     # daily steps
N = int(T/dt)  # number of steps

# Initialize rates array
r = np.zeros(N)
r[0] = r0

# Simulate
for t in range(1, N):
    dr = a * (b - r[t-1]) * dt + sigma * np.sqrt(dt) * np.random.normal()
    r[t] = r[t-1] + dr

# Plot
plt.figure(figsize=(10,6))
plt.plot(np.linspace(0, T, N), r, label='Vasicek short rate simulation')
plt.axhline(b, color='red', linestyle='--', label='Long-term mean (b)')
plt.xlabel('Time (years)')
plt.ylabel('Interest Rate')
plt.title('Vasicek Model Simulation of Short Rates')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
