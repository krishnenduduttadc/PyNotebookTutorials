## Time Value of Money (TVM)

The Time Value of Money is the idea that money available now is worth more than the same amount in the future due to its potential earning capacity.

### Key Formulas:
- **Future Value (FV):**  
  \( FV = PV \times (1 + r)^n \)
  
- **Present Value (PV):**  
  \( PV = \frac{FV}{(1 + r)^n} \)

Where:  
- `PV` is present value  
- `FV` is future value  
- `r` is interest rate per period  
- `n` is number of periods


```python
import numpy as np
import matplotlib.pyplot as plt
# Time Value of Money calculations
def future_value(pv, r, n):
    return pv * (1 + r) ** n

def present_value(fv, r, n):
    return fv / (1 + r) ** n

# Example
pv = 1000  # Initial investment
r = 0.05   # 5% interest
n = np.arange(1, 21)  # Periods from 1 to 20

# Calculate FV
fv = future_value(pv, r, n)

# Plotting
plt.figure(figsize=(10,6))
plt.plot(n, fv, marker='o', label='Future Value')
plt.title('Future Value Over Time at 5% Interest')
plt.xlabel('Years')
plt.ylabel('Future Value')
plt.grid(True)
plt.legend()
plt.show()
```
