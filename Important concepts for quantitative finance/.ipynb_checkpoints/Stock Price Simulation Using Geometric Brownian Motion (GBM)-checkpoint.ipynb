# ðŸ“‰ Geometric Brownian Motion (GBM) â€” Stock Price Simulation

Stock prices are often modeled using **Geometric Brownian Motion (GBM)**, which assumes:

- Returns are normally distributed.
- Prices are continuous and follow a **log-normal** distribution.
- Volatility and drift are constant over time.

### ðŸ”§ GBM Stochastic Differential Equation (SDE)

\[
dS_t = \mu S_t \,dt + \sigma S_t \,dW_t
\]

Where:
- \( S_t \): Stock price at time \( t \)
- \( \mu \): Drift (expected return)
- \( \sigma \): Volatility
- \( dW_t \): Random term from Brownian motion

### âœ… Discretized Formula for Simulation

\[
S_t = S_0 \times \exp\left( (\mu - \frac{1}{2}\sigma^2)t + \sigma W_t \right)
\]

Where \( W_t \) is simulated as cumulative standard normal random variables scaled by \( \sqrt{dt} \).


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = 100        # Initial stock price
mu = 0.07       # Annual expected return
sigma = 0.2     # Annual volatility
T = 1.0         # Time horizon (1 year)
dt = 0.01       # Time step size
N = int(T / dt) # Number of steps
t = np.linspace(0, T, N)  # Time vector

# Simulate Brownian Motion
dW = np.random.normal(0, np.sqrt(dt), N)  # Random normal increments
W = np.cumsum(dW)                         # Cumulative sum = Brownian path

# GBM formula
S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)

# Plot
plt.figure(figsize=(10,6))
plt.plot(t, S, label='Simulated Stock Price')
plt.title('Stock Price Simulation using Geometric Brownian Motion')
plt.xlabel('Time (Years)')
plt.ylabel('Stock Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
