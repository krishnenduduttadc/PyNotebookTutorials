Types of Costs: Fixed, Variable, Semi-Variable
Concept:

Fixed: Do not change with output (e.g., rent)

Variable: Change with output (e.g., raw material)

Semi-variable: Partially fixed + variable (e.g., electricity)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

units = np.arange(0, 101, 10)
fixed_cost = 1000
variable_cost_per_unit = 20
semi_variable_cost = 500 + (10 * units)

df_cost = pd.DataFrame({
    'Units': units,
    'Fixed Cost': fixed_cost,
    'Variable Cost': variable_cost_per_unit * units,
    'Semi-Variable Cost': semi_variable_cost
})

df_cost.plot(x='Units', y=['Fixed Cost', 'Variable Cost', 'Semi-Variable Cost'], kind='line', marker='o', title='Cost Types vs Units')
plt.ylabel("Cost in ")
plt.grid(True)
plt.show()
```

Cost Classification: Direct vs Indirect
Concept:

Direct: Directly traceable to product (e.g., raw material, labor)

Indirect: Not directly traceable (e.g., electricity, rent)

```python
cost_items = pd.DataFrame({
    'Item': ['Raw Material', 'Direct Labor', 'Factory Rent', 'Electricity', 'Supervisor Salary'],
    'Cost': [5000, 3000, 2000, 1000, 1500],
    'Type': ['Direct', 'Direct', 'Indirect', 'Indirect', 'Indirect']
})
cost_items
```

Cost Sheet Preparation
Concept: Tabular summary of all costs to determine total cost and profit.

```python
# Cost Sheet Calculation
prime_cost = 5000 + 3000  # Direct material + Direct labor
factory_overhead = 2000 + 1000 + 1500
total_cost = prime_cost + factory_overhead
selling_price = 13000
profit = selling_price - total_cost

cost_sheet = pd.DataFrame({
    'Component': ['Prime Cost', 'Factory Overhead', 'Total Cost', 'Selling Price', 'Profit'],
    'Amount': [prime_cost, factory_overhead, total_cost, selling_price, profit]
})
cost_sheet
```

Break-Even Analysis
Concept: No profit, no loss point (Fixed Costs / (Selling Price - Variable Cost))

```python
selling_price = 50
variable_cost = 30
fixed_cost = 1000

break_even_units = fixed_cost / (selling_price - variable_cost)

x = np.arange(0, 100)
total_revenue = selling_price * x
total_costs = fixed_cost + variable_cost * x

plt.plot(x, total_revenue, label="Revenue")
plt.plot(x, total_costs, label="Total Cost")
plt.axvline(x=break_even_units, color='red', linestyle='--', label=f'Break Even: {break_even_units:.0f} units')
plt.title("Break-Even Analysis")
plt.xlabel("Units Sold")
plt.ylabel("")
plt.legend()
plt.grid(True)
plt.show()
```

Marginal Costing
Concept: Cost of producing one additional unit = Variable cost

```python
contribution_per_unit = selling_price - variable_cost
total_contribution = contribution_per_unit * 100
total_profit = total_contribution - fixed_cost

pd.DataFrame({
    'Metric': ['Contribution/Unit', 'Total Contribution (100 units)', 'Fixed Cost', 'Profit'],
    'Amount': [contribution_per_unit, total_contribution, fixed_cost, total_profit]
})
```
