Business Descriptive Analytics (EDA)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample sales data
data = {
    'Region': ['North', 'South', 'East', 'West']*25,
    'Sales': abs(np.random.normal(100000, 30000, 100)),
    'Profit': abs(np.random.normal(10000, 5000, 100)),
    'Category': np.random.choice(['A', 'B', 'C'], 100)
}
df = pd.DataFrame(data)

# Descriptive Stats
print(df.describe())

# Visualization - Sales by Region
sns.boxplot(x='Region', y='Sales', data=df)
plt.title("Sales Distribution by Region")
plt.show()
```

KPI Dashboarding

Correlation & Hypothesis Testing

```python
import numpy as np
import scipy.stats as stats

corr = df[['Sales','Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Hypothesis testing
corr_coef, p_value = stats.pearsonr(df['Sales'], df['Profit'])
print(f"Correlation Coefficient: {corr_coef}, P-Value: {p_value}")
```

Demand Forecasting (Time Series)

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Simulated time series data
date_rng = pd.date_range(start='1/1/2022', periods=100, freq='D')
df_ts = pd.DataFrame(date_rng, columns=['date'])
df_ts['sales'] = abs(100000 + 1000*np.arange(100) + np.random.normal(0, 10000, 100))
df_ts.set_index('date', inplace=True)

# Holt-Winters model
model = ExponentialSmoothing(df_ts['sales'], trend="add", seasonal=None)
model_fit = model.fit()
forecast = model_fit.forecast(10)

df_ts['sales'].plot(label='Observed', figsize=(10,6))
forecast.plot(label='Forecast', color='red')
plt.title("Sales Forecast")
plt.legend()
plt.show()
```

Optimization Models
Concept: Linear programming for resource allocation

```python
from scipy.optimize import linprog

# Objective: Maximize profit: 40x + 30y
# Constraints: 
# 1. Labor: 2x + y ≤ 100
# 2. Material: x + y ≤ 80

c = [-40, -30]  # Negative for maximization
A = [[2, 1], [1, 1]]
b = [100, 80]

result = linprog(c, A_ub=A, b_ub=b)
print(result)
```

Strategy-level Analytics (Top 1% Level)
6️⃣ Customer Segmentation (Clustering)
Concept: Segment customers for targeted strategies

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Prepare features
X = df[['Sales','Profit']]

# KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
df['Segment'] = kmeans.labels_

# PCA for 2D visualization
pca = PCA(2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:,0], X_pca[:,1], c=df['Segment'])
plt.title("Customer Segmentation via Clustering")
plt.show()
```

Pricing Simulation & Game Theory (Strategic Thinking)

```python
# Two competitors pricing game
prices = np.arange(50, 150, 10)
profit_matrix = np.zeros((len(prices), len(prices)))

for i, my_price in enumerate(prices):
    for j, competitor_price in enumerate(prices):
        market_share = 1 / (1 + np.exp(my_price - competitor_price))
        profit_matrix[i, j] = my_price * market_share * 1000

sns.heatmap(profit_matrix, xticklabels=prices, yticklabels=prices)
plt.title("Profit Matrix vs Competitor Pricing")
plt.xlabel("Competitor Price")
plt.ylabel("My Price")
plt.show()
```
