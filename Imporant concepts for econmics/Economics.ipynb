Supply and Demand

```python
import numpy as np
import matplotlib.pyplot as plt

# Define price points
prices = np.linspace(1, 20, 100)

# Demand decreases as price increases (Law of Demand)
demand = 100 - 4 * prices

# Supply increases as price increases (Law of Supply)
supply = 20 + 3 * prices

# Find equilibrium (where demand = supply)
equilibrium_idx = np.argmin(np.abs(demand - supply))
eq_price = prices[equilibrium_idx]
eq_quantity = demand[equilibrium_idx]

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(prices, demand, label='Demand Curve', color='blue')
plt.plot(prices, supply, label='Supply Curve', color='orange')
plt.plot(eq_price, eq_quantity, 'ro', label='Equilibrium Point')
plt.title("Supply and Demand")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.legend()
plt.grid(True)
plt.annotate(f'Equilibrium\nPrice: {eq_price:.2f}\nQuantity: {eq_quantity:.0f}',
             xy=(eq_price, eq_quantity), xytext=(eq_price+2, eq_quantity+10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()
```

Price Elasticity of Demand

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated prices
prices = np.array([5, 6, 7, 8, 9])
# Simulated quantities sold at those prices
quantities = np.array([80, 70, 60, 50, 40])

# Calculate % changes
percent_change_price = (prices[1:] - prices[:-1]) / prices[:-1]
percent_change_quantity = (quantities[1:] - quantities[:-1]) / quantities[:-1]

# Elasticity = % change in quantity / % change in price
elasticity = percent_change_quantity / percent_change_price

# Plot elasticity
mid_prices = (prices[1:] + prices[:-1]) / 2

plt.figure(figsize=(8, 5))
plt.plot(mid_prices, elasticity, marker='o')
plt.title("Price Elasticity of Demand")
plt.xlabel("Price (Midpoint)")
plt.ylabel("Elasticity")
plt.axhline(y=-1, color='r', linestyle='--', label='Unit Elastic (-1)')
plt.legend()
plt.grid(True)
plt.show()

# Print result
for i, e in enumerate(elasticity):
    print(f"Elasticity between ${prices[i]} and ${prices[i+1]}: {e:.2f}")
```

Break-even Analysis

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fixed_costs = 10000
price_per_unit = 50
variable_cost_per_unit = 20

# Range of units
units = np.arange(0, 1000)

# Total Revenue = Price × Quantity
revenue = price_per_unit * units

# Total Cost = Fixed + Variable × Quantity
total_cost = fixed_costs + variable_cost_per_unit * units

# Break-even point
break_even_units = fixed_costs / (price_per_unit - variable_cost_per_unit)
break_even_revenue = price_per_unit * break_even_units

# Plot
plt.figure(figsize=(10, 6))
plt.plot(units, revenue, label='Total Revenue', color='green')
plt.plot(units, total_cost, label='Total Cost', color='red')
plt.axvline(x=break_even_units, linestyle='--', color='black', label='Break-even Point')
plt.title("Break-even Analysis")
plt.xlabel("Units Sold")
plt.ylabel("USD")
plt.legend()
plt.grid(True)
plt.annotate(f'Break-even:\n{break_even_units:.0f} units\n${break_even_revenue:.0f}',
             xy=(break_even_units, break_even_revenue),
             xytext=(break_even_units + 100, break_even_revenue - 5000),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()

# Print result
print(f"Break-even quantity: {break_even_units:.0f} units")
print(f"Break-even revenue: ${break_even_revenue:.2f}")
```

Profit Maximization

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
price_per_unit = 60
variable_cost_per_unit = 30
fixed_cost = 5000
units = np.arange(0, 300)

# Calculations
revenue = price_per_unit * units
total_cost = fixed_cost + variable_cost_per_unit * units
profit = revenue - total_cost

# Optimal units for max profit
max_profit_idx = np.argmax(profit)
max_profit_units = units[max_profit_idx]
max_profit_value = profit[max_profit_idx]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(units, revenue, label="Total Revenue", color="green")
plt.plot(units, total_cost, label="Total Cost", color="red")
plt.plot(units, profit, label="Profit", color="blue")
plt.axvline(x=max_profit_units, linestyle='--', color='gray', label="Max Profit Point")
plt.title("Profit Maximization")
plt.xlabel("Units Sold")
plt.ylabel("USD")
plt.legend()
plt.grid(True)
plt.annotate(f'Max Profit: ${max_profit_value:.0f}\nat {max_profit_units} units',
             xy=(max_profit_units, max_profit_value),
             xytext=(max_profit_units + 20, max_profit_value - 2000),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()

# Print
print(f"Maximum profit of ${max_profit_value:.2f} occurs at {max_profit_units} units.")
```

Market Structures

```python
import numpy as np
import matplotlib.pyplot as plt

# Quantity
q = np.linspace(1, 100, 100)

# Perfect Competition: Price is fixed
price_pc = 50
revenue_pc = price_pc * q
mr_pc = np.full_like(q, price_pc)  # constant marginal revenue

# Monopoly: Downward sloping demand
price_m = 100 - 0.5 * q
revenue_m = price_m * q
mr_m = np.gradient(revenue_m, q)

# Marginal cost (same in both cases)
mc = 10 + 0.2 * q  # increasing marginal cost

# Plot
plt.figure(figsize=(12, 6))

# Perfect Competition
plt.subplot(1, 2, 1)
plt.plot(q, mr_pc, label='MR (PC)', color='blue')
plt.plot(q, mc, label='MC', color='red')
plt.title("Perfect Competition")
plt.xlabel("Quantity")
plt.ylabel("Price / Cost")
plt.legend()
plt.grid(True)

# Monopoly
plt.subplot(1, 2, 2)
plt.plot(q, mr_m, label='MR (Monopoly)', color='blue')
plt.plot(q, mc, label='MC', color='red')
plt.title("Monopoly")
plt.xlabel("Quantity")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Opportunity Cost

```python
import numpy as np
import matplotlib.pyplot as plt

# Total available resources (e.g., hours, capital)
resources = 100

# Resource per unit
res_A = 2  # Product A needs 2 units per item
res_B = 5  # Product B needs 5 units per item

# Max units possible
max_A = resources // res_A
max_B = resources // res_B

# Generate production possibilities
units_A = np.arange(0, max_A + 1)
units_B = (resources - res_A * units_A) / res_B  # how many B units remain

# Plot
plt.figure(figsize=(8, 6))
plt.plot(units_A, units_B, label='Production Possibility Frontier', color='purple')
plt.title("Opportunity Cost: Product A vs Product B")
plt.xlabel("Units of Product A")
plt.ylabel("Units of Product B")
plt.grid(True)
plt.fill_between(units_A, units_B, alpha=0.1, color='purple')
plt.annotate("Opportunity Cost Curve\n(Tradeoff between A & B)",
             xy=(20, 6), xytext=(30, 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()
```

Time Value of Money (NPV & IRR)

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy_financial import npv, irr

# Cash flow: [Initial investment, yearly returns...]
cash_flows = [-10000, 3000, 3500, 4000, 4500]
years = np.arange(len(cash_flows))

# NPV calculation at different discount rates
rates = np.linspace(0.01, 0.25, 100)
npvs = [npv(r, cash_flows) for r in rates]

# IRR
project_irr = irr(cash_flows)

# Plot NPV vs Discount Rate
plt.figure(figsize=(10, 6))
plt.plot(rates * 100, npvs, label='NPV Curve')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(project_irr * 100, color='red', linestyle='--', label=f'IRR ≈ {project_irr*100:.2f}%')
plt.title("NPV vs Discount Rate")
plt.xlabel("Discount Rate (%)")
plt.ylabel("NPV (₹)")
plt.grid(True)
plt.legend()
plt.show()

# Output
print(f"IRR for the project: {project_irr*100:.2f}%")
```

Inflation & Purchasing Power

```python
import numpy as np
import matplotlib.pyplot as plt

# Initial value
initial_value = 100000

# Years
years = np.arange(0, 21)

# Inflation rates
inflation_rates = [0.03, 0.05, 0.07]  # 3%, 5%, 7%

# Calculate real value over time
for rate in inflation_rates:
    future_value = initial_value / ((1 + rate) ** years)
    plt.plot(years, future_value, label=f'{int(rate*100)}% Inflation')

# Plot
plt.title("Purchasing Power Over Time")
plt.xlabel("Years")
plt.ylabel("Real Value of ₹100,000")
plt.legend()
plt.grid(True)
plt.show()
```

Cash Flow Forecasting

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Simulate 12 months
months = np.arange(1, 13)

# Revenue grows over time
revenue = 50000 + (months * 3000)

# Fixed costs (e.g. rent, salaries)
fixed_cost = 30000

# Variable costs: 40% of revenue
variable_cost = 0.4 * revenue

# Total costs
total_cost = fixed_cost + variable_cost

# Net cash flow
cash_flow = revenue - total_cost

# Starting cash reserve
starting_cash = 20000
ending_cash = np.cumsum(cash_flow) + starting_cash

# Plot
plt.figure(figsize=(12, 6))
plt.plot(months, revenue, label="Revenue", color='green')
plt.plot(months, total_cost, label="Total Cost", color='red')
plt.plot(months, ending_cash, label="Ending Cash", color='blue', marker='o')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Monthly Cash Flow Forecast")
plt.xlabel("Month")
plt.ylabel("Amount (₹)")
plt.legend()
plt.grid(True)
plt.show()

# Show as table
df = pd.DataFrame({
    "Month": months,
    "Revenue": revenue.astype(int),
    "Total Cost": total_cost.astype(int),
    "Net Cash Flow": cash_flow.astype(int),
    "Ending Cash": ending_cash.astype(int)
})
print(df)
```

 Unit Economics: CAC vs LTV

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate CAC and LTV for different months
months = np.arange(1, 13)
cac = np.random.normal(1200, 100, len(months))  # Avg ₹1200, some variation
ltv = np.linspace(2500, 5000, len(months))       # Improving LTV over time

# Plot
plt.figure(figsize=(10, 6))
plt.plot(months, cac, label='CAC (₹)', color='red', marker='o')
plt.plot(months, ltv, label='LTV (₹)', color='green', marker='o')
plt.axhline(0, color='gray', linestyle='--')
plt.fill_between(months, ltv, cac, where=(ltv>cac), interpolate=True, color='green', alpha=0.2, label='Profitable Months')
plt.fill_between(months, ltv, cac, where=(ltv<=cac), interpolate=True, color='red', alpha=0.2, label='Unprofitable Months')
plt.title("Unit Economics: CAC vs LTV Over Time")
plt.xlabel("Month")
plt.ylabel("₹")
plt.legend()
plt.grid(True)
plt.show()

# Print insights
profitable_months = np.sum(ltv > cac)
print(f"Profitable months: {profitable_months} out of {len(months)}")
print(f"Target: Maintain LTV ≥ 3 × CAC → ideal LTV should be ≥ ₹{3 * np.mean(cac):.2f}")
```

Risk & Sensitivity Analysis

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
units_sold = 1000
fixed_cost = 20000

# Price and cost ranges
prices = np.linspace(40, 80, 20)
costs = np.linspace(20, 60, 20)

# Create profit matrix
profit_matrix = np.zeros((len(prices), len(costs)))

for i, price in enumerate(prices):
    for j, cost in enumerate(costs):
        total_revenue = price * units_sold
        total_cost = fixed_cost + cost * units_sold
        profit_matrix[i, j] = total_revenue - total_cost

# Plot sensitivity heatmap
plt.figure(figsize=(10, 8))
plt.contourf(costs, prices, profit_matrix, levels=50, cmap="RdYlGn")
plt.colorbar(label='Profit (₹)')
plt.title("Profit Sensitivity to Price and Cost")
plt.xlabel("Variable Cost per Unit (₹)")
plt.ylabel("Selling Price per Unit (₹)")
plt.grid(True)
plt.show()
```

Break-Even with Uncertainty (Monte Carlo Simulation)

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation count
n = 1000

# Random variables
np.random.seed(42)  # reproducibility
price = np.random.normal(60, 5, n)
cost = np.random.normal(40, 4, n)
units = np.random.normal(1000, 100, n)
fixed_cost = 20000

# Calculate profit for each simulation
revenue = price * units
variable_cost = cost * units
profit = revenue - variable_cost - fixed_cost

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(profit, bins=40, color='skyblue', edgecolor='black')
plt.axvline(np.mean(profit), color='red', linestyle='--', label=f'Mean Profit: ₹{np.mean(profit):.0f}')
plt.title("Monte Carlo Simulation: Profit Distribution")
plt.xlabel("Profit (₹)")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
plt.show()

# Probability of loss
prob_loss = np.mean(profit < 0)
print(f"Probability of Loss: {prob_loss * 100:.2f}%")
```

Startup Valuation Techniques

```python
import numpy as np
import matplotlib.pyplot as plt

# Forecasted annual free cash flows (in ₹ lakhs)
cash_flows = [0, 5, 15, 25, 35, 45]  # Year 0 to Year 5
discount_rate = 0.12
years = np.arange(len(cash_flows))

# DCF Calculation
dcf_values = [cf / (1 + discount_rate) ** yr for yr, cf in enumerate(cash_flows)]
npv = sum(dcf_values)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(years, dcf_values, color='dodgerblue', alpha=0.8)
plt.title("Startup DCF Valuation (Discounted Cash Flows)")
plt.xlabel("Year")
plt.ylabel("Discounted Cash Flow (₹ lakhs)")
plt.xticks(years)
plt.grid(True)
plt.annotate(f"NPV ≈ ₹{npv:.2f} lakhs", xy=(3, 15), xytext=(1, 30),
             arrowprops=dict(arrowstyle='->', color='black'))
plt.show()

print(f"Startup Valuation (NPV): ₹{npv:.2f} lakhs")
```
